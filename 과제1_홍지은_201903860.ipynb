```python
import urllib.request
from bs4 import BeautifulSoup
import re
import pandas as pd

result = []

for sido in range(1, 18):
    for gu in range(1, 26):
        try:
            Kyochon_url = 'https://www.kyochon.com/shop/domestic.asp?txtsearch=&sido1=%s&sido2=%s' % (sido, gu)
            print(Kyochon_url)
            html = urllib.request.urlopen(Kyochon_url)
            soupKyo = BeautifulSoup(html, 'html.parser')
            tag_tbody = soupKyo.find('tbody')  
            tr_tags = tag_tbody.find_all("tr")

            for tr in tr_tags:
                store = tr.find_all("td")
                name = store[0].get_text().strip() 
                addr = store[1].get_text().strip() 

                addr_list = re.split(r'\s(?=\S+$)', addr)
                if len(addr_list) == 3:
                    sido_str, gu_str, detail_addr = addr_list
                else:
                    sido_str = "" 
                    gu_str = ""
                    detail_addr = ""

                result.append([name, sido_str, gu_str, detail_addr])

        except:
            pass

# 데이터프레임 생성
df = pd.DataFrame(result, columns=['상호명', '시도', '구', '상세주소'])

# csv 파일로 저장
df.to_csv('kyochon_stores.csv', index=False)
```
